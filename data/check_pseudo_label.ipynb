```python

import os
import numpy as np
import tqdm
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import random
import tqdm
import random
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
import json

from pycocotools.mask import encode, decode
import numpy as np


sa1b_dir = "/home/dchenbs/workspace/datasets/sa1b"
output_dir = "/home/dchenbs/workspace/datasets/sa1b_directsam/res_1800px_thr.03_output" # pseudo boundary label generated by DirectSAM
# output_dir = "/home/dchenbs/workspace/datasets/sa1b_directsam/res_3600px_thr.005_output" # pseudo boundary label generated by DirectSAM

# list all json files in output_dir including subdirectories
image_paths = []
json_files = []
for root, dirs, files in os.walk(output_dir):
    for file in files:
        if file.endswith(".json"):
            json_file = os.path.join(root, file)
            json_files.append(json_file)
            image_paths.append(json_file.replace(output_dir, sa1b_dir).replace(".json", ".jpg"))

print(f"Found {len(json_files)} json files")


for _ in range(5):

    i = random.randint(0, len(json_files) - 1)
    json_file = json_files[i]
    image_path = image_paths[i]

    with open(json_file, "r") as f:
        results = json.load(f)

    image = Image.open(image_path).resize((1800, 1800))
    image = np.array(image)

    boundary = decode(results)

    plt.figure(figsize=(30, 15))
    plt.subplot(1, 2, 1)
    plt.title(f'Input Image: {image.shape}')
    plt.imshow(image)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(boundary, cmap='PuBuGn')
    plt.title(f'DirectSAM Prediction')
    plt.axis("off")

    plt.show()
        # break
```
